```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="gemma:2b")
)
```

```python
embeddings
```

```python
r1=embeddings.embed_documents(
    [
        "Alpha is the first character of greek alphabet"
        "Beta is the second character of greek alphabet"

    ]
)
```

```python
r1
```

```python
r1[0]
```

```python
embeddings.embed_query("What is the first character of greek alphabet?")
```

```python
text="This is my first test"
query_result=embeddings.embed_query(text)
query_result
```
